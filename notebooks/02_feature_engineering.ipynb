## 2. Feature Engineering (juste pour faire jolie pour l'instant)

The feature engineering phase aims to transform raw data into meaningful and informative variables that improve the performance of machine learning models. This step is crucial because the quality of the features directly impacts the predictive power of the models.

### 2.1 Objectives

The main objectives of feature engineering are:

- Improve model performance  
- Capture underlying patterns in the data  
- Reduce noise and irrelevant information  
- Prepare data in a format suitable for modeling  

### 2.2 Data Cleaning

Before creating new features, the dataset was cleaned to ensure consistency and reliability.

This includes:

- Handling missing values (removal or imputation)  
- Removing duplicates  
- Correcting inconsistent formats  
- Converting data types when necessary  

Clean data is essential to avoid bias and incorrect model behavior.

### 2.3 Feature Transformation

Several transformations were applied to make the data more suitable for modeling:

- **Scaling / Normalization**: ensures features are on comparable scales  
- **Log transformations**: reduce skewness in highly asymmetric distributions  
- **Encoding categorical variables**:
  - One-hot encoding  
  - Label encoding  

These transformations help models learn more effectively, especially those sensitive to feature scale.

### 2.4 Feature Creation

New variables were created to extract additional information from existing data.

Examples include:

- **Temporal features** (if time data is present):
  - Year, month, day, season  
- **Lag features** (for time-series or forecasting tasks):
  - Previous values of the target variable  
- **Interaction features**:
  - Combination of multiple variables  
- **Aggregated features**:
  - Averages, sums, or rolling statistics  

These features help capture hidden patterns not directly visible in the raw data.

### 2.5 Feature Selection

Not all features contribute positively to model performance. Therefore, feature selection was performed to retain only the most relevant variables.

Methods used include:

- Correlation analysis  
- Removal of highly correlated features (multicollinearity)  
- Feature importance from models  
- Domain knowledge  

This step reduces overfitting and improves model interpretability.

### 2.6 Handling Outliers

Outliers can negatively affect model performance. They were handled using:

- Detection via statistical methods (e.g., IQR, z-score)  
- Removal or transformation when necessary  

However, care was taken to ensure that important extreme values were not mistakenly removed.

### 2.7 Final Feature Set

After all transformations, the final dataset consists of a refined set of features that are:

- Clean  
- Relevant  
- Scaled and encoded  
- Informative for prediction  

This dataset is then used as input for the modeling phase.

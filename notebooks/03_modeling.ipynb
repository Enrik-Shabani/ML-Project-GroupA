## 3. Modeling (juste pour faire jolie pour l'instant)

The modeling phase aims to develop predictive models that address the research question while ensuring good generalization performance on unseen data. Given the objective of the project, the problem is formulated as a supervised machine learning task.

### 3.1 Modeling Strategy

To ensure robust results, multiple models with varying levels of complexity were tested. This approach allows comparison between simple baseline methods and more advanced algorithms.

The modeling process follows these steps:

1. Define the target variable and input features  
2. Split the dataset into training and test sets  
3. Train a baseline model  
4. Train multiple machine learning models  
5. Perform hyperparameter tuning  
6. Evaluate model performance  
7. Compare models and select the best one  

### 3.2 Train-Test Split

The dataset is divided into a training set and a test set. The training set is used to fit the models, while the test set is used for final evaluation.

This ensures that performance reflects the model’s ability to generalize rather than memorize the training data. When necessary, cross-validation is applied to improve robustness.

### 3.3 Baseline Model

A baseline model is established to provide a reference point.

- **Regression**: predicting the mean value  
- **Classification**: predicting the majority class  

This allows evaluation of whether more complex models provide meaningful improvement.

### 3.4 Selected Models

Several supervised learning algorithms are tested:

- **Linear / Logistic Regression**: simple and interpretable baseline  
- **Decision Tree**: captures non-linear relationships  
- **Random Forest**: reduces variance and improves accuracy  
- **Gradient Boosting (e.g., XGBoost)**: strong predictive performance  
- **Support Vector Machine (optional)**  
- **Neural Networks (optional)**  

These models provide a balance between interpretability and performance.

### 3.5 Hyperparameter Tuning

Hyperparameters are optimized using techniques such as grid search or randomized search with cross-validation.

Examples include:

- Number of trees  
- Maximum depth  
- Learning rate  
- Regularization parameters  

This step improves model performance beyond default settings.

### 3.6 Evaluation Metrics

Evaluation metrics depend on the task:

#### Regression:
- MAE (Mean Absolute Error)  
- RMSE (Root Mean Squared Error)  
- R² score  

#### Classification:
- Accuracy  
- Precision  
- Recall  
- F1-score  
- ROC-AUC  

Using multiple metrics provides a complete evaluation of performance.

### 3.7 Model Comparison

Models are compared based on performance, robustness, and interpretability. The best model is selected based on the overall trade-off between accuracy and simplicity.

### 3.8 Feature Importance and Interpretability

Feature importance analysis is conducted to understand the model:

- Coefficients (linear models)  
- Feature importance (tree-based models)  
- SHAP values (advanced interpretability)  

This helps identify key drivers of predictions.

### 3.9 Overfitting and Limitations

To reduce overfitting, techniques such as cross-validation and proper data splitting are used. However, model performance remains dependent on data quality and feature selection.
